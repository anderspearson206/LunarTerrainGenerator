# Lunar Highlands Terrain Generator

Example Usage notebook

```python
from lunar_terrain_generator import LTGen
import numpy as np

```

First, instantiate the terrain generator

```python
mygenerator = LTGen()
```

before terrain can be generated, crater profiles need to be precomputed storing these profiles vastly speeds up terrain generation in exchange for higher memory usage. The low_memory option attempts to limit memoryt usage. The low memory parameter can be turned off if RAM is not a limitation

```python
mygenerator.generate_crater_dict(low_memory=True)
```

To generate terrain, use the .gen_highland() function, or .gen_maria()

```python

example_terrain = mygenerator.gen_highland()
ex_maria = mygenerator.gen_maria()
```

To display the terrain as a DTM (digital terrain model), use .display_DTM()

```python

mygenerator.display_DTM(example_terrain, "example highland terrain")
mygenerator.display_DTM(ex_maria, "example maria terrain")
```
